## Functioanl API 작성법에 더 익숙해야한다.
- 앞 처리 로직의 **결과**가 이어지는 처리 로직의 **입력 데이터**로 주어지는 **Chain**형태의 프로그래밍 구현 로직에 적합

### Functional API 작성법

```python
from tensorflow.keras.layers import Input, Flatten, Dense
from tensorflow.keras.models import Model

INPUT_SIZE = 28

input_tensor = Input(shape=(INPUT_SIZE, INPUT_SIZE))                # input
x = Flatten()(input_tensor)                                         #  28 by 28
x = Dense(100, activation='relu', name='Dense_1')(x)
x = Dense(30, activation='relu', name='Dense_2')(x)
output_tensor = Dense(10, activation='softmax', name='Dense_3')(x)  # output

model = Model(inputs=input_tensor, outputs=output_tensor)

model.summary()
```

### Sequential 작성법

```python
from tensorflow.keras.layers import Input, Flatten, Dense
from tensorflow.keras.models import Model, Sequential

INPUT_SIZE = 28

model = Sequential([
    Flatten(input_shape=(INPUT_SIZE, INPUT_SIZE)),
    Dense(100, activation='relu', name='Dense_1'),
    Dense(30, activation='relu', name='Dense_2'),
    Dense(10, activation='softmax', name='Dense_3')
])

model.summary()
```
